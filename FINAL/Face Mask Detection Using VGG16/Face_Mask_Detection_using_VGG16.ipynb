Importing Libraries

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten, BatchNormalization,Dropout
import matplotlib.pyplot as plt
from keras import callbacks
import numpy as np
```

Image Augmentation

```python
from keras.preprocessing.image import ImageDataGenerator

batch_size=32

train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=40,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        'Face Mask Dataset/Train',
        target_size=(224, 224),
        batch_size=batch_size,
        class_mode='binary') 

validation_generator = val_datagen.flow_from_directory(
        'Face Mask Dataset/Validation',
        target_size=(224, 224),
        batch_size=batch_size,
        class_mode='binary')
```

```python
# Generators
train_ds = keras.utils.image_dataset_from_directory(
    directory='Face Mask Dataset/Train',
    labels='inferred',
    batch_size=32,
    image_size=(224,224),

)

validation_ds = keras.utils.image_dataset_from_directory(
    directory='Face Mask Dataset/Validation',
    labels='inferred',
    batch_size=32,
    image_size=(224,224),

)
```

```python
class_names = train_ds.class_names
class_names
```

```python
# Looking into the sample data
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

```python
for image_batch, labels_batch in train_generator:
    print(image_batch.shape)
    print(labels_batch.shape)
    break
```

```python
# Normalize the image to bring it down to a size between (0,1)
def process_image(image, label):
    image = tf.cast(image/255. , tf.float32)
    return image, label

train_ds = train_ds.map(process_image)
validation_ds = validation_ds.map(process_image)
```

Creating a base model

```python
# Create CNN Model
model = Sequential()

model.add(Conv2D(32, kernel_size=(3,3), padding='valid', activation='relu', input_shape=(224,224,3)))
model.add(MaxPooling2D(pool_size=(2,2), strides=2, padding='valid'))

model.add(Conv2D(64, kernel_size=(3,3), padding='valid', activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2), strides=2, padding='valid'))

model.add(Conv2D(128, kernel_size=(3,3), padding='valid', activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2), strides=2, padding='valid'))

model.add(Flatten())

model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Fit the model
history = model.fit(train_ds,epochs=20, validation_data=validation_ds)
```

Plotting the Graph

```python
# Comparing accuracy
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.grid(True)
plt.plot(history.history['accuracy'], label='Training accuracy', color='orange')
plt.plot(history.history['val_accuracy'], label='Validation accuracy', color='blue')
plt.legend()


# Comparing loss
plt.subplot(1,2,2)
plt.grid(True)
plt.plot(history.history['loss'], label='Training Loss', color='orange')
plt.plot(history.history['val_loss'], label='Validation Loss', color='blue')
plt.legend()
plt.show()
```

Creating a VGG16 Model

```python
# Create a VGG model
from keras.applications.vgg16 import VGG16

conv_base = VGG16(
    weights='imagenet',
    include_top = False,
    input_shape=(224,224,3)
)
```

```python
model_vgg = Sequential()
model_vgg.add(conv_base)
model_vgg.add(Flatten())
model_vgg.add(Dense(256,activation='relu'))
model_vgg.add(Dense(1,activation='sigmoid'))

model_vgg.summary()
```

```python
conv_base.trainable = False

model_vgg.summary()
```

```python
model_vgg.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history_aug = model_vgg.fit(
        train_generator,
        epochs=1,
        validation_data=validation_generator)
```

```python
def draw_label(img,text,pos,bg_color):
  text_size = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX,1, cv2.FILLED)

  end_x = pos[0] + text_size[0][0] + 2
  end_y = pos[1] + text_size[0][1] - 2

  cv2.rectangle(img, pos, (end_x,end_y), bg_color, cv2.FILLED)
  cv2.putText(img, text, pos, cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,0), 1, cv2.LINE_AA)
```

```python
def detect_face_mask(img):
    y_pred = model_vgg.predict(img.reshape(1,224,224,3), verbose=0)
    #y_pred = model_vgg.predict(img, verbose=0)
  #np.where(y_pred[0][0] > 0.5, 1,0)
  #y_pred = (model.predict(img.reshape(1,224,224,3), verbose=0) > 0.5).astype("int32")
    if y_pred[0][0] >= 0.5:
        return 1
    else:
        return 0
```

```python
sample_1 = cv2.imread('Face Mask Dataset/Test/WithoutMask/1001.png')
sample_1 = cv2.resize(sample_1,(224,224))
sample_mask_img = np.reshape(sample_1,[1,224,224,3])
sample_mask_img = sample_mask_img/255.0
```

```python
detect_face_mask(sample_mask_img)
```

```python
cap = cv2.VideoCapture(0)
while True:
     ret, frame = cap.read()
        
     #Call the detection method
     img = cv2.resize(frame, (224,224),3)
     y_pred = detect_face_mask(img)
   
     if y_pred == 0:
         draw_label(frame, "Mask", (30,30), (0,255,0))
     else:
         draw_label(frame, "No Mask", (30,30), (0,0,255))

     cv2.imshow("window", frame)
    
     if cv2.waitKey(1) & 0xFF == ord('x'):
         break
cap.release()
cv2.destroyAllWindows()
```
